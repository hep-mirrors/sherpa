```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import lpmv  

#Discrete Quadrature on a Spherical Grid

G0 = np.zeros((13, 13))
H0 = np.zeros((13, 13))

g_data = [
    [-29496.6, -1586.3], [ -2396.6, 3026.1, 1668.6], [1340.1, -2326.2, 1231.9, 634.0],
    [912.6, 808.9, 166.7, -357.1, 89.4], [-230.9, 357.2, 200.3, -141.1, -163.0, -7.8],
    [72.8, 68.6, 76.0, -141.4, -22.8, 13.2, -77.9], [80.5, -75.1, -4.7, 45.3, 13.9, 10.4, 1.7, 4.9],
    [24.4, 8.1, -14.5, -5.6, -19.3, 11.5, 10.9, -14.1, -3.7], [5.4, 9.4, 3.4, -5.2, 3.1, -12.4, -0.7, 8.4, -8.5, -10.1],
    [-2.0, -6.3, 0.9, -1.1, -0.2, 2.5, -0.3, 2.2, 3.1, -1.0, -2.8], [3.0, -1.5, -2.1, 1.7, -0.5, 0.5, -0.8, 0.4, 1.8, 0.1, 0.7, 3.8],
    [-2.2, -0.2, 0.3, 1.0, -0.6, 0.9, -0.1, 0.5, -0.4, -0.4, 0.2, -0.8, 0.0]
]

h_data = [
    [0, 4944.4], [0, -2707.7, -576.1], [0, -160.2, 251.9, -536.6],
    [0, 286.4, -211.2, 164.3, -309.1], [0, 44.6, 188.9, -118.2, 0.0, 100.9],
    [0, -20.8, 44.1, 61.5, -66.3, 3.1, 55.0], [0, -57.9, -21.1, 6.5, 24.9, 7.0, -27.7, -3.3],
    [0, 11.0, -20.0, 11.9, -17.4, 16.7, 7.0, -10.8, 1.7], [0, -20.5, 11.5, 12.8, -7.2, -7.4, 8.0, 2.1, -6.1, 7.0],
    [0, 2.8, -0.1, 4.7, 4.4, -7.2, -1.0, -3.9, -2.0, -2.0, -8.3], [0, 0.2, 1.7, -0.6, -1.8, 0.9, -0.4, -2.5, -1.3, -2.1, -1.9, -1.8],
    [0, 0.0, -0.9, 0.3, 2.1, -2.5, 0.5, 0.6, 0.0, 0.3, -0.9, -0.2, 0.9]
]

for i, row in enumerate(g_data):
    G0[i+1, :len(row)] = row
for i, row in enumerate(h_data):
    H0[i+1, :len(row)] = row

# --- 2. CONSTANTS ---
RE = 6371.0
M_PHI = 0.001 # Screening factor (1/km)


def calcIntegral(r_obs):
    # Observation point (At the North Pole)
    theta_obs = 0.0
    phi_obs = 0.0
    
    # sampling points for the integral, might increase later for better resolution. 
    N_r, N_ct, N_p = 30, 30, 30
    
    r_p_grid = np.linspace(3000, 6500, N_r)
    ct_p_grid = np.linspace(-1, 1, N_ct)
    phi_p_grid = np.linspace(0, 2 * np.pi, N_p)
    
    dr = (6500 - 3000) / N_r
    dct = 2 / N_ct
    dp = (2 * np.pi) / N_p
    
    integral_val = 0.0
    
    for rp in r_p_grid: # picking a shell at a specific radius
        for ctp in ct_p_grid: #picking a horizontal slice of that shell (the latitude?) like cos(theta) part. 
            
            #Instead of a third loop, we calculate all longitudinal points, 0 to 2pi, at once
            stp = np.sqrt(max(0, 1 - ctp**2)) # recall : 1 - cos^2 theta = sin^2 theta
            
            # Geometry: cos(gamma) calculation
            cos_gamma = (np.sin(theta_obs) * stp * np.cos(phi_p_grid - phi_obs) + 
                         np.cos(theta_obs) * ctp)
            
            rho = np.sqrt(( r_obs**2 + rp**2 - 2 * r_obs * rp * cos_gamma)) # (r^2 + rp^2 - 2r rp cos(gamma), where 
            #cos(gamma) is cos(theta)cos(theta') + sin(theta)sin(theta')cos(phi'-phi)
            
            kernel = (rp**2 * np.exp(-M_PHI * rho)) / rho # our integrand 
            
            # Summation over n and m
            harmonic_sum = np.zeros_like(phi_p_grid)
            for n in range(1, 13):
                # Legendre polynomials
                p_nm = lpmv(np.arange(n + 1), n, ctp) # this is legendre polyn. 
                
                term_n = (RE / rp)**(n + 2) * (n + 1)
                
                # Trigonometric part for all m at once
                m_vals = np.arange(n + 1)[:, None]
                cos_mphi = np.cos(m_vals * phi_p_grid)
                sin_mphi = np.sin(m_vals * phi_p_grid)
                
                trig_sum = (G0[n, :n+1][:, None] * cos_mphi + 
                            H0[n, :n+1][:, None] * sin_mphi)
                
                harmonic_sum += term_n * np.sum(p_nm[:, None] * trig_sum, axis=0)
            
            integral_val += np.sum(kernel * harmonic_sum) # where the summing happens. 
            
    return (1 / (2 * np.pi)) * integral_val * dr * dct * dp

r_axis = np.linspace(6600, 10000, 100)
print(f"Calculating {len(r_axis)} points.")

phi_results = []
for i, r in enumerate(r_axis):
    val = calcIntegral(r)
    phi_results.append(val)
   ##print(f"Progress: {i+1}/{len(r_axis)} points calculated.")

plt.figure(figsize=(10, 6))
plt.plot(r_axis, phi_results, 'b-o', linewidth=2)
plt.title(r'Calculated Particular Solution $\Phi_x$ vs Radius $r$')
plt.xlabel('Distance r (km)')
plt.ylabel('Particular Soln')
plt.grid(True, linestyle='--')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import lpmv  
from scipy.special import factorial, lpmv

#Discrete Quadrature on a Spherical Grid

G0 = np.zeros((13, 13))
H0 = np.zeros((13, 13))

g_data = [
    [-29496.6, -1586.3], [ -2396.6, 3026.1, 1668.6], [1340.1, -2326.2, 1231.9, 634.0],
    [912.6, 808.9, 166.7, -357.1, 89.4], [-230.9, 357.2, 200.3, -141.1, -163.0, -7.8],
    [72.8, 68.6, 76.0, -141.4, -22.8, 13.2, -77.9], [80.5, -75.1, -4.7, 45.3, 13.9, 10.4, 1.7, 4.9],
    [24.4, 8.1, -14.5, -5.6, -19.3, 11.5, 10.9, -14.1, -3.7], [5.4, 9.4, 3.4, -5.2, 3.1, -12.4, -0.7, 8.4, -8.5, -10.1],
    [-2.0, -6.3, 0.9, -1.1, -0.2, 2.5, -0.3, 2.2, 3.1, -1.0, -2.8], [3.0, -1.5, -2.1, 1.7, -0.5, 0.5, -0.8, 0.4, 1.8, 0.1, 0.7, 3.8],
    [-2.2, -0.2, 0.3, 1.0, -0.6, 0.9, -0.1, 0.5, -0.4, -0.4, 0.2, -0.8, 0.0]
]

h_data = [
    [0, 4944.4], [0, -2707.7, -576.1], [0, -160.2, 251.9, -536.6],
    [0, 286.4, -211.2, 164.3, -309.1], [0, 44.6, 188.9, -118.2, 0.0, 100.9],
    [0, -20.8, 44.1, 61.5, -66.3, 3.1, 55.0], [0, -57.9, -21.1, 6.5, 24.9, 7.0, -27.7, -3.3],
    [0, 11.0, -20.0, 11.9, -17.4, 16.7, 7.0, -10.8, 1.7], [0, -20.5, 11.5, 12.8, -7.2, -7.4, 8.0, 2.1, -6.1, 7.0],
    [0, 2.8, -0.1, 4.7, 4.4, -7.2, -1.0, -3.9, -2.0, -2.0, -8.3], [0, 0.2, 1.7, -0.6, -1.8, 0.9, -0.4, -2.5, -1.3, -2.1, -1.9, -1.8],
    [0, 0.0, -0.9, 0.3, 2.1, -2.5, 0.5, 0.6, 0.0, 0.3, -0.9, -0.2, 0.9]
]

for i, row in enumerate(g_data):
    G0[i+1, :len(row)] = row
for i, row in enumerate(h_data):
    H0[i+1, :len(row)] = row

# --- 2. CONSTANTS ---
RE = 6371.0
M_PHI = 0.001 # Screening factor (1/km)


def calcIntegral(r_obs):
    # Observation point (At the North Pole)
    theta_obs = 0.0
    phi_obs = 0.0
    
    # sampling points for the integral, might increase later for better resolution. 
    N_r, N_ct, N_p = 51, 51, 51
    
    r_p_grid = np.linspace(2890, 6500, N_r)
    ct_p_grid = np.linspace(-1, 1, N_ct)
    phi_p_grid = np.linspace(0, 2 * np.pi, N_p)
    
    dr = (6500 - 3000) / N_r
    dct = 2 / N_ct
    dp = (2 * np.pi) / N_p
    
    integral_val = 0.0
    
    for rp in r_p_grid: # picking a shell at a specific radius
        i = 0
        for ctp in ct_p_grid: #picking a horizontal slice of that shell (the latitude?) like cos(theta) part. 
            
            #Instead of a third loop, we calculate all longitudinal points, 0 to 2pi, at once
            stp = np.sqrt(max(0, 1 - ctp**2)) # recall : 1 - cos^2 theta = sin^2 theta
            
            # Geometry: cos(gamma) calculation
            cos_gamma = (np.sin(theta_obs) * stp * np.cos(phi_p_grid - phi_obs) + 
                         np.cos(theta_obs) * ctp)
            
            rho = np.sqrt(( r_obs**2 + rp**2 - 2 * r_obs * rp * cos_gamma)) # (r^2 + rp^2 - 2r rp cos(gamma), where 
            #cos(gamma) is cos(theta)cos(theta') + sin(theta)sin(theta')cos(phi'-phi)
            
            kernel = (rp**2 * np.exp(-M_PHI * rho)) / rho # our integrand 
            
            # Summation over n and m
            harmonic_sum = np.zeros_like(phi_p_grid)
            
            for n in range(1, 13):
                m_range = np.arange(n + 1)
                
                # 1. Compute the Schmidt/Normalization factor for all m at once
                # Note: for m=0, the '2' in the numerator is often omitted in 
                # some Schmidt conventions. Check if you need: (2 if m > 0 else 1)
                norm_factor = ((-1)**m_range) * np.sqrt(2 * factorial(n - m_range) / factorial(n + m_range))
                
                # 2. Legendre polynomials (n+1,)
                p_nm_raw = lpmv(m_range, n, ctp) 
                
                p_nm = p_nm_raw * norm_factor
                
                # 4. Radial and Trig parts (same as before)
                term_n = (RE / rp)**(n + 2) * (n + 1)
                
                m_vals = m_range[:, None]
                cos_mphi = np.cos(m_vals * phi_p_grid)
                sin_mphi = np.sin(m_vals * phi_p_grid)
                
                # 5. Summing over m
                g_m = G0[n, :n+1, None]
                h_m = H0[n, :n+1, None]
                
                trig_sum = (g_m * cos_mphi) + (h_m * sin_mphi)
                
                # harmonic_sum update
                harmonic_sum += term_n * np.sum(p_nm[:, None] * trig_sum, axis=0)
            
            integral_val += np.sum(kernel * harmonic_sum) # where the summing happens. 
        i = i+1
    return (1 / (2 * np.pi)) * integral_val * dr * dct * dp * slope[i]

r_axis = np.linspace(6600, 10000, 100)
print(f"Calculating {len(r_axis)} points.")

phi_results = []
for i, r in enumerate(r_axis):
    val = calcIntegral(r)
    phi_results.append(val)
   ##print(f"Progress: {i+1}/{len(r_axis)} points calculated.")

plt.figure(figsize=(10, 6))
plt.plot(r_axis, phi_results, 'b-o', linewidth=2)
plt.title(r'Calculated Particular Solution $\Phi_x$ vs Radius $r$')
plt.xlabel('Distance r (km)')
plt.ylabel('Particular Soln')
plt.grid(True, linestyle='--')
plt.show()
```
