```python
#Import functions and set design parameters
import numpy as np 
from matplotlib import pyplot as plt
from matplotlib import rc
import glob as glob

from functions import *

plt.rcParams["axes.grid"] = False

rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)

#### Design properties ####
fScaling = 1.5 #Decides overall scaling of figures, 1 should be optimal
fWidth = fScaling*4.7747
fHeight = fScaling*3.81976
fontSize = 10.5*fScaling
fontSizeLegend = 9.5*fScaling

path_end = "_s*" + random_seed_end + ".dat"

path_start_1D = "../HI_Analysis/"
path_start_2D = "../HI_Analysis2D/"
```

## 1. Accept/reject overdraw<a id="1-mg2"></a>
Shows the average amount of times (per event) that the accept reject probability > 1, i.e. when the original overestimate has been too low. Also shows the average value of the accept/reject ratio for the cases where it is > 1.

```python
plt.figure(figsize=(fWidth,fHeight))
plt.title(r'$\langle N_{overdraw} \rangle$ for scatter/merge and split, per event, runcard ' + random_seed_end, fontsize=fontSize)
    
histo_name = "N_overdraw_split"
x, y, y_mean, y_err_1, y_err_2, N_groups = Plot2D(path_start_1D + histo_name + path_end)
if N_groups > 0:
    plot_label_1 = r'$\langle N_{\mathrm{overdraw, split}} \rangle$ = ' + r'{:.3f}'.format(y_mean) + r'$\pm$' + r'{:.3f}'.format(y_err_2)+r' (Error combined data)'
    plot_label_2 = r'$\langle N_{\mathrm{overdraw, split}} \rangle$ = ' + r'{:.3f}'.format(y_mean) + r'$\pm$' + r'{:.3f}'.format(y_err_1)+r' (Error between groups)'
    if np.sum(y) > 0:
        y = y/np.sum(y)
    plt.plot(x, y, '-', color='tab:blue', linewidth=1.5, label = plot_label_1 + '\n' + plot_label_2)
else:
    print("Could not find any data for histograms: " + histo_name)
    
histo_name = "N_overdraw_scatter_merge"
x, y, y_mean, y_err_1, y_err_2, N_groups = Plot2D(path_start_1D + histo_name + path_end)
if N_groups > 0:
    plot_label_1 = r'$\langle N_{\mathrm{overdraw, scatter/merge}} \rangle$ = ' + r'{:.3f}'.format(y_mean) + r'$\pm$' + r'{:.3f}'.format(y_err_2)+r' (Error combined data)'
    plot_label_2 = r'$\langle N_{\mathrm{overdraw, scatter/merge}} \rangle$ = ' + r'{:.3f}'.format(y_mean) + r'$\pm$' + r'{:.3f}'.format(y_err_1)+r' (Error between groups)'
    if np.sum(y) > 0:
        y = y/np.sum(y)
    plt.plot(x, y, '--', color='forestgreen', linewidth=1.5, label = plot_label_1 + '\n' + plot_label_2)
else:
    print("Could not find any data for histograms: " + histo_name)
    
plt.ylabel(r'Probability', fontsize=fontSize)
plt.xlabel(r'$N_{overdraw}$', fontsize=fontSize)
plt.legend(loc="upper right",prop={"size":0.9*fontSizeLegend})
plt.tight_layout()


plt.figure(figsize=(fWidth,fHeight))
plt.title(r'Accept/reject ratio for scatter/merge and split, runcard ' + random_seed_end, fontsize=fontSize)
    
histo_name = "val_overdraw_split"
x, y, y_mean, y_err_1, y_err_2, N_groups = Plot2D(path_start_1D + histo_name + path_end)
if N_groups > 0:
    plot_label_1 = r'$\langle$ od. ratio split $\rangle$ = ' + r'{:.3f}'.format(y_mean) + r'$\pm$' + r'{:.3f}'.format(y_err_2)+r' (Error combined data)'
    plot_label_2 = r'$\langle$ od. ratio split $\rangle$ = ' + r'{:.3f}'.format(y_mean) + r'$\pm$' + r'{:.3f}'.format(y_err_1)+r' (Error between groups)'
    if np.sum(y) > 0:
        y = y/np.sum(y)
    plt.plot(x, y, '-', color='tab:blue', linewidth=1.5, label = plot_label_1 + '\n' + plot_label_2)
else:
    print("Could not find any data for histograms: " + histo_name)
    
histo_name = "val_overdraw_scatter_merge"
x, y, y_mean, y_err_1, y_err_2, N_groups = Plot2D(path_start_1D + histo_name + path_end)
if N_groups > 0:
    plot_label_1 = r'$\langle$ od. ratio scatter/merge $\rangle$ = ' + r'{:.3f}'.format(y_mean) + r'$\pm$' + r'{:.3f}'.format(y_err_2)+r' (Error combined data)'
    plot_label_2 = r'$\langle$ od. ratio scatter/merge $\rangle$ = ' + r'{:.3f}'.format(y_mean) + r'$\pm$' + r'{:.3f}'.format(y_err_1)+r' (Error between groups)'
    if np.sum(y) > 0:
        y = y/np.sum(y)
    plt.plot(x, y, '--', color='forestgreen', linewidth=1.5, label = plot_label_1 + '\n' + plot_label_2)
else:
    print("Could not find any data for histograms: " + histo_name)
    
plt.ylabel(r'Probability', fontsize=fontSize)
plt.xlabel(r'accept/reject ratio', fontsize=fontSize)
plt.legend(loc="upper right",prop={"size":0.9*fontSizeLegend})
plt.tight_layout()
```
